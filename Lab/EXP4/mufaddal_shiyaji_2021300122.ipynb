```r

data <- read.csv("Property_stolen_and_recovered.csv")


head(data)
```

```r
library(ggplot2)

# Bar chart for Cases_Property_Stolen by Area_Name
# The bar chart shows the number of cases of property stolen in different Indian states and
# territories. Maharashtra has the highest number of cases, followed by Delhi and Andhra Pradesh.


ggplot(data, aes(x = Area_Name, y = Cases_Property_Stolen)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  coord_flip() + 
  labs(title = "Cases of Property Stolen by Area", x = "Area", y = "Cases Property Stolen")
```

```r
#pie chart
#The pie chart shows the top 5 areas by property recovered. Maharashtra has the highest
# percentage of property recovered, followed by Madhya Pradesh.

top5_data <- head(data[order(-data$Value_of_Property_Recovered), ], 5)


pie(top5_data$Value_of_Property_Recovered, 
    labels = top5_data$Area_Name, 
    col = rainbow(5),
    main = "Top 5 Areas by Property Recovered")
```

```r
#histogram
#he histogram shows the distribution of cases of property recovered. Most cases involve
# the recovery of a relatively small amount of property, with fewer cases involving the recovery of larger amounts.

ggplot(data, aes(x = Value_of_Property_Stolen)) +
  geom_histogram(fill = "blue", color = "black", alpha = 0.7) +
  xlim(0, 1.5e09) + ylim(0,500)+ 
  labs(title = "Distribution of Cases_Property_Recovered",
       x = "Cases_Property_Recovered",
       y = "Count")
```

```r
# Time line chart for a hypothetical dataset (assuming time data)
# The line chart shows the trend of property stolen over time in different Indian states and 
#territories. Most states experienced a general increase in the value of property stolen between
# 2002 and 2010, with some fluctuations.

ggplot(data, aes(x = Year, y = Value_of_Property_Stolen, group = Area_Name, color = Area_Name)) +
  geom_line() +
  labs(title = "Trend of Property Stolen Over Time", x = "Year", y = "Value of Property Stolen")
```

```r
# Scatter plot for Value of Property Stolen vs Recovered
# The scatter plot shows the relationship between the value of property stolen and the value of property recovered in different 
# Indian states and territories.
ggplot(data, aes(x = Value_of_Property_Stolen, y = Value_of_Property_Recovered)) +
  geom_point(aes(color = Area_Name, size = Cases_Property_Stolen)) + ylim(0,2.5e09)+
  labs(title = "Property Stolen vs Property Recovered", x = "Value of Property Stolen", y = "Value of Property Recovered")
```
